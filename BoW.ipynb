```python
# 학습, 평가 데이터 나누기
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords as nl
```

```python
d = pd.read_csv("weatherReports.csv")
d.head()
```

```python
d=d.dropna() # NaN 이 있는 데이터 행은 모두 제거
```

```python
d=d[(d['event_type']=='Thunderstorm Wind') | (d['event_type']=='Hail')]
```

```python
d.loc[(d['event_type']=='Thunderstorm Wind'), ['event_type']] = 0
d.loc[(d['event_type']=='Hail'), ['event_type']] = 1
```

```python
d
```

```python
train_data, test_data = train_test_split(d, test_size=0.3, shuffle=True)
# 학습 데이터 7 : 평가 데이터 3 나누기
train = train_data.to_numpy()
test = test_data.to_numpy()
```

```python
stopwords = stopwords.words('english')
stopwords += [str(i) for i in range(1000)]
```

```python
vector = CountVectorizer(stop_words=stopwords)
```

```python
vector.fit_transform(train[:, 10]).toarray()
```

```python
vector.vocabulary_
```

```python
W = np.array(np.zeros_like(vector.transform([train[1, 10]]).toarray())[0])
b = 0.0 # 절편과 기울기 초기화
epochs = 1000
lr = 0.1
```

```python
def H(x):
    z = np.dot(W, x) + b # ax + b
    h = 1/(1+(1/np.exp(z)))
    if h>0.5:
        return 1
    else:
        return 0
```

```python
from tqdm import tqdm_notebook

for i in tqdm_notebook(range(epochs)):
    for j in range(len(train)):
        x = np.array(vector.transform([train[j, 10]]).toarray()[0])
        y = train[j, 3]
        h = H(x)
        W = W + (lr*(y-h)*x)
        b = b + (lr*(y-h))
```

```python
count = 0
for i in range(len(test)):
    x = np.array(vector.transform([test[i, 10]]).toarray()[0])
    if train[i, 3] == H(x):
        count += 1
        
print(count/len(test))
```
