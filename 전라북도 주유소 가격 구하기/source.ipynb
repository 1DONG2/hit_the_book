```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome("chromedriver.exe")
driver.get("https://www.opinet.co.kr/searRgSelect.do")
```

```python
si_list_raw = driver.find_element_by_xpath("""//*[@id="SIGUNGU_NM0"]""")
si_list = si_list_raw.find_elements_by_tag_name('option')
si_list
```

```python
si_names = [option.get_attribute('value') for option in si_list]
si_names.remove('')
si_names
```

```python
from selenium.webdriver.support.ui import Select
element = Select(driver.find_element_by_id("SIGUNGU_NM0"))
element.select_by_value(si_names[0])
```

```python
xpath= """//*[@id="searRgSelect"]"""
element_sel_si = driver.find_element_by_xpath(xpath).click()
```

```python
xpath= """//*[@id="glopopd_excel"]"""
element_sel_excel = driver.find_element_by_xpath(xpath).click()
```

```python
import time
from selenium.webdriver.support.ui import Select
from tqdm import tqdm_notebook

for si in tqdm_notebook(si_names):
    element = Select(driver.find_element_by_id("SIGUNGU_NM0"))
    element.select_by_value(si)
    time.sleep(2)

    xpath= """//*[@id="searRgSelect"]"""
    element_sel_si = driver.find_element_by_xpath(xpath).click()
    time.sleep(1)

    xpath= """//*[@id="glopopd_excel"]"""
    element_sel_excel = driver.find_element_by_xpath(xpath).click()
    time.sleep(1)
```

```python
import pandas as pd
from glob import glob

glob('전라도주유소/지역*.xls')
```

```python
stations_files = glob('전라도주유소/지역*.xls') # 파일에 있는 파일명들을 불러와 리스트로 저장
stations_files
```

```python
tmp_raw = []

for file_name in stations_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)

station_raw = pd.concat(tmp_raw)
```

```python
station_raw.info()
```

```python
station_raw.head()
```

```python
stations = pd.DataFrame({'Oil_store':station_raw['상호'],
'주소':station_raw['주소'],
'가격':station_raw['휘발유'],
'셀프':station_raw['셀프여부'],
'상호':station_raw['상표'],}) # 이름 바꾸기 및 간략히 추리기
stations.head()
```

```python
stations['시/군'] = [eachAddress.split()[1] for eachAddress in stations['주소']] 
# 스플릿으로 한칸 건너띄고 추출하면 그거 나오자나 그래서 그걸로 리스트 함축시키자.
stations
```

```python
stations['시/군'].unique()
```

```python
# 난데없이 익산시여야하는데 익산이라고 되는 귀찮은 상황이 발생할때
stations[stations['시/군']=='익산']
```

```python
stations.loc[stations['시/군']=='익산', '시/군'] = '익산시' # 이제 잘 되었다. ^^
stations['시/군'].unique()
```

```python
stations[stations['가격']=='-'] 
# 가격이 갱신이 되지 않은 괘씸한 주유소들이다.
```

```python
stations = stations[stations['가격'] != '-'] # - 로 적혀진 주유소는 빼고 다시 저장한다.
stations.head()
```

```python
stations['가격'] = [float(value) for value in stations['가격']] # 가격이 문자형이기 때문에 계산이나 통계내기위해 숫자로 바꿔줌
```
