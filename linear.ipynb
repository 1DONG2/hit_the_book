```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
exdata = pd.read_csv("ex1data1.txt", names=['x', 'y'])
exdata
```

```python
plt.scatter(exdata.x, exdata.y)
plt.show()
```

```python
W=0.0 # 기울기
b=0.0 # 절편
Irate=0.001 # 학습율
epochs=50000 # 반복 횟수
n=len(exdata.x) # 입력 데이터의 개수
```

```python
for i in range(epochs): # 50000번 만큼 경사하강법 반복
    Y_pred = W*exdata.x+b # 직선방정식
    # 해당 w, b를 미분
    dW =(2/n)*sum(exdata.x*(Y_pred-exdata.y))
    db = (2/n)*sum(Y_pred-exdata.y)
    # 미분한 값을 학습률에 곱해 지속 업데이트
    W = W-Irate*dW
    b = b-Irate*db
    # 100번에 한번씩 결과 출력
    if i % 100 == 0:
        print("반복 횟수",i,"기울기: ", round(W,4),"절편: ", round(b,4))
```
